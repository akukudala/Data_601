```python
# import pandas lib as pd
import pandas as pd
import plotly
import plotly.express as px
 
# read by default 1st sheet of an excel file
df = pd.read_excel("Cleaned_data601.xlsx")
 
df.sample(5)
```

```python
df = df.dropna()
```

```python
df.sample(5)
```

```python
fig = px.scatter(
    df,
    title="Comparison between Unemployment Rate and Graduation Rate ",
    x="Unemp", # Unemployment Rate
    y="County Value%", # Graduation Rate
    # size="County Value",
    color="Year",
    template="plotly_dark",
    labels={"Unemployment dec ":"Unemployment dec ", "County Value":"County Value"}
)
# fig.update_traces(textposition='top left')
# Years 
fig.update_layout(showlegend= False)
fig.show()
```

```python
fig = px.scatter(
    df,
    title="Per Pupil Spending Vs Unemployment Rate",
    x="Unemp", # Unemployment Rate
    y="PPCSTOT",
    # size="County Value",
    color="Year",
    template="plotly_dark",
    labels={"Unemployment dec 2015":"Unemployment dec ", "County Value":"County Value"}
)
# fig.update_traces(textposition='top left')

fig.update_layout(showlegend= False)
fig.show()
```

```python
df=df."Locale*".str.replace("(::).*","")
```

```python
df["General County Type"]= df["Locale*"].str.replace("(:).*","")
```

```python
df.sample(5)
```

```python

df2 = df.groupby(['General County Type']).agg(
    Hours_Mean = ('County Value%', 'mean'))

df2
```

```python
df3=df2.reset_index()
```

```python
df3
```

```python
fig = px.scatter(
    df3,
    title="General County Type grouped",
    x="General County Type",
    y="Hours_Mean",
    size="Hours_Mean",
    color="General County Type",
    template="plotly_dark",
    labels={"Unemployment dec 2015":"Unemployment dec 2015", "County Value":"County Value"}
)
# fig.update_traces(textposition='top left')

fig.update_layout(showlegend= False)
fig.show()
```

```python
fig = px.scatter(
    df,
    title="Unemployment Rate Vs Graduation Rate color coded by General County Type",
    x="Unemp",
    y="County Value%",
    # size="County Value",
    color="General County Type",
    template="plotly_dark",
    labels={"Unemployment dec ":"Unemployment dec ", "County Value":"County Value"}
)
# fig.update_traces(textposition='top left')

fig.update_layout(showlegend= False)
fig.show()
```

```python
fig = px.scatter(
    df,
    title="Per Pupil Spending Vs Graduation Rate",
    x="PPCSTOT",
    y="County Value%",
    # size="County Value",
    color="General County Type",
    template="plotly_dark",
    labels={"Unemployment dec ":"Unemployment dec ", "County Value":"County Value"}
)
# fig.update_traces(textposition='top left')

fig.update_layout(showlegend= False)
fig.show()
```
